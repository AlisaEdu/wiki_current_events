```python
# load data
from google.colab import drive
from google.colab import files

import pandas as pd
import requests 
import json

drive.mount('/content/drive')

df = pd.read_csv("drive/My Drive/Diploma/second_try/Code/add_category/views_events_1_year_edits_1_day.csv")
df.head(2)
```

```python
def get_topic(title):
  try:
    url = f'https://wikipedia-topic.wmcloud.org/api/v1/topic?lang=en&title={title}'
    r = requests.get(url).text
    r = json.loads(r)['results'][0] # score, topic
    a, b = r['score'], r['topic']
  except:
    a, b = -1, -1
  return title, a, b

a = "Prime_Minister_of_Haiti"
get_topic(a)
```

```python
def get_region(title):
  try:
    url = f'https://wiki-region.wmcloud.org/api/v1/region?lang=en&titles={title}'
    r = requests.get(url).text
    r = json.loads(r)[0]['regions'][0]
    a, b, c = ['region'], r['subcontinent'], r['continent']
  except:
    a, b, c = -1, -1, -1
  return title, a, b, c

a = "Prime_Minister_of_Haiti"
get_region(a)
```

```python
def get_quality(title):
  try:
    url = f'https://misalignment.wmcloud.org/api/v1/quality-article?lang=en&title={title}'
    r = requests.get(url).text
    r = json.loads(r)
    a, b = r['class'], r['quality']
  except:
    a, b = -1, -1
  return title, a, b

a = "Prime_Minister_of_Haiti"
get_quality(a)
```
